```python
%sql
-- 1. Tabla Silver con un CTAS
CREATE OR REPLACE TABLE fire_risk_project.02_silver.silver_grid_master AS
SELECT 
    cast(cell_id as STRING) as grid_id, 
    cast(latitude as DOUBLE) as latitude_centroid,
    cast(longitude as DOUBLE) as longitude_centroid,
    current_timestamp() as fecha_procesamiento -- Por las dudas

FROM fire_risk_project.00_landing.aux_grid_master
WHERE latitude IS NOT NULL AND longitude IS NOT NULL;

-- Optimizaci칩n de tabla Delta para lecturas r치pidas (Z-Ordering por geo)
OPTIMIZE fire_risk_project.02_silver.silver_grid_master 
ZORDER BY (latitude_centroid, longitude_centroid);
```
