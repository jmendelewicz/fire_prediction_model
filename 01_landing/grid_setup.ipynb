```python
#Librerías

import numpy as np
import pandas as pd
```

```python
# Box de extracción: Sierras de Córdoba

LAT_MIN, LAT_MAX = -32.5, -30.5
LON_MIN, LON_MAX = -65.0, -64.2
RESOLUCION = 0.1 

# Puntos en la grilla. Referencia para la extracción de datos climáticos. 

lats = np.arange(LAT_MIN, LAT_MAX, RESOLUCION)
lons = np.arange(LON_MIN, LON_MAX, RESOLUCION)

grid_data = []
for lat in lats:
    for lon in lons:
        grid_lat = round(lat, 2)
        grid_lon = round(lon, 2)
        # ID Único. Para joins
        cell_id = f"{grid_lat}_{grid_lon}"
        grid_data.append({"cell_id": cell_id, "latitude": grid_lat, "longitude": grid_lon})

df_grid = spark.createDataFrame(pd.DataFrame(grid_data))
df_grid.write.mode("overwrite").saveAsTable("fire_risk_project.00_landing.aux_grid_master")

print(f"Grilla con {len(grid_data)} puntos de monitoreo.")
display(df_grid)
```
